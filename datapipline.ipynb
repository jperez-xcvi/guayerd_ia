```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
current_dir = Path.cwd()
normal_data = current_dir.joinpath("data", "normal_true")

#nombres de los archivos
clientes_file = "clientes.csv"
ventas_file = "ventas.csv"
productos_file = "productos.csv"
detalle_ventas_file = "detalle_ventas.csv"
ciudades_file = "argentina_ciudades.csv"

#cargar los datos
clientes = pd.read_csv(normal_data.joinpath(clientes_file))
ventas = pd.read_csv(normal_data.joinpath(ventas_file))
productos = pd.read_csv(normal_data.joinpath(productos_file))
detalle_ventas = pd.read_csv(normal_data.joinpath(detalle_ventas_file))
ciudades = pd.read_csv(normal_data.joinpath(ciudades_file))
```

```python
# conexion a la base de datos
load_dotenv()
host = os.getenv("DB_HOST")
puerto = os.getenv("DB_PORT")
user = os.getenv("DB_USER")
password = os.getenv("DB_PASSWORD")
base_datos = os.getenv("DB_NAME")

# Crear conexión
engine = create_engine(f'postgresql://{user}:{password}@{host}:{puerto}/{base_datos}')

# Verificar que la conexión funciona
with engine.connect() as conn:
    print("✅ Conexión exitosa a PostgreSQL")
```

```python
# Enviar DataFrames a PostgreSQL
# if_exists='replace' sobrescribe la tabla, usa 'append' para agregar datos

ciudades.to_sql('argentina_ciudad', con=engine, if_exists='append', index=False)
print("✅ Tabla 'argentina_ciudad' cargada")

productos.to_sql('producto', con=engine, if_exists='append', index=False)
print("✅ Tabla 'producto' cargada")

clientes.to_sql('cliente', con=engine, if_exists='append', index=False)
print("✅ Tabla 'cliente' cargada")

ventas.to_sql('venta', con=engine, if_exists='append', index=False)
print("✅ Tabla 'venta' cargada")

detalle_ventas.to_sql('detalle_venta', con=engine, if_exists='append', index=False)
print("✅ Tabla 'detalle_venta' cargada")

print("\n✅ ¡Todos los datos han sido cargados exitosamente!")
```
